```python
TRAIN_PATH = '../data/train'
TEST_PATH = '../data/test'
```

```python
train_walk = os.walk(TRAIN_PATH)
main_dir_path, _, rle_path = next(train_walk)
dir_im_path, _, train_im_path = next(train_walk)
dir_mask_path, _, train_mask_path = next(train_walk)
```

```python
main_dir_path, dir_im_path, dir_mask_path
```

```python
train_df = pd.DataFrame()
train_df['id'] = [im_p.split('.')[0] for im_p in train_im_path]
train_df['im_paths'] = [os.path.join(dir_im_path, im_p) for im_p in train_im_path]
train_df['mask_paths'] = [os.path.join(dir_mask_path, m_p) for m_p in train_mask_path]
rle_df = pd.read_csv(os.path.join(main_dir_path, rle_path[0]))
train_df = train_df.merge(rle_df, on='id', how='left')
```
